# Desktop Env 请求样例

```python
import requests

USER_TOKEN = "kYHj5v9LmQp3XcR2sWnB7zTq8yFgK1J"
ENV_URL = "http://112.125.88.107:4999"


def get_server_list():
    """
    获取指定USER TOKEN下的所有env server
    """
    try:
        headers = {
            "Authorization": USER_TOKEN,
            "Content-Type": "application/json"
        }
        response = requests.get(
            f"{ENV_URL}/server/list", headers=headers
        )
        if response.status_code == 200:
            # print("获取服务器列表成功")
            return response.json()["data"]
        else:
            print(f"获取服务器列表失败: {response.text}")
            return []
    except Exception as e:
        print(f"获取服务器列表异常: {e}")
        return []
    
def release_server_env(server_id):
    """
    根据server_id释放env
    """ 
    try:
        headers = {
            "Authorization": USER_TOKEN,
            "Content-Type": "application/json"
        }
        response = requests.post(
            f"{ENV_URL}/server/release/{server_id}", headers=headers
        )
        if response.status_code == 200:
            print(f"释放服务器成功: {server_id}")

            return True
        else:
            print(f"释放服务器失败: {response.text}")
            return False
    except Exception as e:
        print(f"释放服务器异常: {e}")
        return False
```
